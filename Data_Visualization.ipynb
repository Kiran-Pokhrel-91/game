```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# imort datasets
titanicdf = sns.load_dataset('titanic')
titanicdf.info()
```

```python
titanicdf
```

```python
titanicdf.isnull().sum()
```

# Bar plot

```python
print(titanicdf.survived.value_counts())
titanicdf.survived.value_counts().plot(kind='bar')
```

```python
print(titanicdf.groupby(['survived','sex'])['survived'].count())
sns.countplot(x="survived",hue="sex",data=titanicdf)
```

# Scatter plot

```python
# plt.scatter(data=titanicdf, x="age", y="fare", c="pclass", cmap='BuGn')
sns.scatterplot(data=titanicdf,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=titanicdf,x="sex",y="fare",hue="pclass")
```

# Hist plot

```python
titanicdf.hist(figsize=(16,10),grid=False)
```

# box plot


```python
sns.boxplot(data=titanicdf,x="sex",y="fare",hue="class")
```

```python
sns.boxenplot(data=titanicdf,x="sex",y="fare",hue="class")
```

```python
g = sns.FacetGrid(titanicdf,hue="survived",col="sex",margin_titles=True,
                  palette={1:"seagreen",0:"gray"})
g = g.map(plt.scatter,"fare","age",edgecolor="w").add_legend()
```
